# Distplots in Seaborn

```python
#Importing the libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

**Loading the Penguins Dataset**

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

**Drawing a Simple distplot**

```python
sns.distplot(penguins.flipper_length_mm,rug = True)
```

**Removing the KDE Plot**

```python
sns.distplot(penguins.flipper_length_mm,hist = False)
```

**Setting the number of bins**

```python
sns.distplot(penguins.flipper_length_mm,bins = 20)
```

**Setting the bin values as array**

```python
sns.distplot(penguins.flipper_length_mm,bins = [160,170,180,190,200,240])
```

**Fitting the data to a particular distribution using norm_hist**

```python
sns.distplot(penguins.flipper_length_mm,norm_hist = True,kde = False)
```

**Styling Rugplots**

```python
sns.distplot(penguins.flipper_length_mm,rug_kws = {'color':'red','height':0.30},rug = True)
```

**Styling Histograms**

```python
sns.distplot(penguins.flipper_length_mm,kde = False,hist_kws = {'color':'blue','alpha':0.2,'histtype':'stepfilled'})
```

**Styling KDE Plots**

```python
sns.distplot(penguins.flipper_length_mm,kde = True,hist = False,kde_kws = {'color':'red','lw':4.0,'alpha':0.9})
```
